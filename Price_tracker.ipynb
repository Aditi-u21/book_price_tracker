```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/book_price_tracker.csv")


df['Timestamp'] = pd.to_datetime(df['Timestamp'])

df['Price']=df['Price'].str.replace("Â£","").astype(float)

rating_map = {
    "One":1,
    "Two":2,
    "Three":3,
    "Four":4,
    "Five":5
}

df['Rating']=df['Rating'].map(rating_map)
df.dtypes
```

```python
#  ðŸ“Š Plot 1: Price Distribution
plt.figure(figsize=(10, 5))
plt.hist(df["Price"], bins=30, color="turquoise", edgecolor="black")
plt.title("Distribution of Book Prices")
plt.xlabel("Price (Â£)")
plt.ylabel("Number of Books")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Get rating counts as a Series (not DataFrame)
rating_counts = df['Rating'].value_counts().sort_index()

# Step 2: Plot bar chart
plt.figure(figsize=(8, 5))
plt.bar(rating_counts.index, rating_counts.values, color="skyblue", edgecolor="black")

# Step 3: Label it
plt.title("Book Rating Distribution")
plt.xlabel("Rating (Stars)")
plt.ylabel("Number of Books")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
